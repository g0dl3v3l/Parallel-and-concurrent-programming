# Experiment 1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def process_file(file_path):
    with open(file_path , 'r') as file:
        content = file.read()

    throughPut = content.split('\n')
    throughPut = throughPut[:len(throughPut)-1]
    rv = [float(i) for i in throughPut]    

    return rv
        
throughPut_filterLock = process_file('./OutputFiles/experiment_1_output_file_bakeryLock.txt')

print(throughPut_filterLock)
throughPut_bakeryLock = process_file('./OutputFiles/experiment_1_output_file_filterLock.txt')


n_values = [2**i for i in range(1, 7)]

plt.figure()
plt.plot(n_values, throughPut_filterLock, marker='o', color='blue', label='FilterLock', linestyle='-')
plt.plot(n_values, throughPut_bakeryLock, marker='o', color='red', label='bakeryLock', linestyle='-')
plt.xlabel('Number of threads')
plt.ylabel('Throughput(ms)')
plt.legend(title='Method')
plt.title('Experiment 1: Throughput Analysis with varying threads')

plt.grid(True)

# Show plot
plt.show()
```
